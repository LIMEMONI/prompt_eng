````python
def generate_interview_question(system_prompt="", user_prompt="", temperature=1, max_tokens=1024):
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {
                "role": "system",
                "content": system_prompt  # 시스템 역할의 메시지로 시스템 프롬프트를 전달합니다.
            },
            {
                "role": "user",
                "content": user_prompt  # 사용자 역할의 메시지로 사용자 프롬프트를 전달합니다.
            }
        ],
        temperature=temperature,  # 온도 설정
        max_tokens=max_tokens,  # 최대 토큰 수 설정
    )

    return completion.choices[0].message.content  # 완성된 메시지 중 첫 번째 메시지의 내용을 반환합니다.
````
